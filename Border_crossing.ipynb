# Border Crossing Analysis


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

BC = pd.read_csv('./Border_Crossing_Entry_Data.csv')

plt.style.use('seaborn-dark')
```

```
Pname = sorted(BC['Port Name'].unique())
Tot = len(Pname)
print(f'There are {Tot} Port of entry to USA from below border sharing states')
BC[['State', 'Port Name']].groupby('State')['Port Name'].nunique().plot(kind='bar')
```

```
BC[ BC['Border'] == 'US-Mexico Border'].groupby('State')['Value'].mean().plot(kind='bar')
BC[ BC['Border'] == 'US-Canada Border'].groupby('State')['Value'].mean().plot(kind='bar')
```
