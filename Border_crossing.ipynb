# Border Crossing Analysis


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

BC = pd.read_csv('./Border_Crossing_Entry_Data.csv')

plt.style.use('seaborn-dark')
```

```
Pname = sorted(BC['Port Name'].unique())
Tot = len(Pname)
print(f'There are {Tot} Port of entry to USA from below border sharing states')
BC[['State', 'Port Name']].groupby('State')['Port Name'].nunique().plot(kind='pie')
```

```
print(f'US Canda Border vs US Mexico Border')
BC[['Border', 'State']].groupby('Border')['State'].nunique().plot(kind='bar')
```

```
BC[ BC['Border'] == 'US-Mexico Border'].groupby('State')['Value'].mean().plot(kind='bar')
BC[ BC['Border'] == 'US-Canada Border'].groupby('State')['Value'].mean().plot(kind='bar')
```

```
BC[['Measure', 'Value']].groupby('Measure')['Value'].nunique().plot(kind='bar')
```

```
BC['Date_as_date'] = pd.to_datetime(BC['Date'])
BC[['Date', 'Date_as_date']]

BC.groupby(BC["Date_as_date"].dt.month)['Value'].count().plot(kind="bar")
plt.ylabel('Number of Vehicles')
plt.xlabel('Month')
plt.title('Crossing per month');
```
